## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
# Tampilan dataset
df
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Tenure',
       'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary', 'Exited']].copy()

df
```

```python
#Menampilkan kolom pada df yg bertipe objek/kategorik

categorical = df.select_dtypes(include=['object']).columns
categorical
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df, columns=['Geography', 'Gender'])
```

```python
# Tampilan df setelah dilakukan One-Hot Encoder data categorical
df
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.drop(columns=['Exited'])  # Mengambil semua kolom kecuali 'Exited' sebagai feature
y = df['Exited']  # Mengambil kolom 'Exited' sebagai target
```

```python
#Lakukan Scaler dan/atau Normalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE

#Inisialisasi MinMaxScaler
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisasi. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)

#X_transform = X.copy()
```

```python
# Tampilan X setelah dilakukan scaler
X_transform
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model 1 : Logistic Regression

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model
    #from sklearn.linear_model import LogisticRegression
    #model1 = LogisticRegression()
    #params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

# Import library yang diperlukan dari scikit-learn
from sklearn.linear_model import LogisticRegression

#START CODE

# Inisialisasi model logistik regresi
model1 = LogisticRegression()
# Buat kumpulan Hyperparameter yg akan diuji
params =  {
    'tol': [0.0001, 0.001, 0.01, 0.1],  # Toleransi (tolerance)
    'C': [0.001, 0.01, 0.1, 1, 10, 100]  # Kebalikan kekuatan regulasi
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model1, # Model yang akan diuji
             param_grid= params, # Kumpulan hyperparameter yang akan diuji
             scoring = 'accuracy', # Metrik evaluasi yang digunakan
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 10-fold cross validation (artinya kita melakukan iterasi model sebanyak 10 kali)
            )

# Melatih model dengan kombinasi hyperparameter terbaik
grid.fit(X_train,y_train)

# Menampilkan hyperparameter terbaik yang ditemukan oleh GridSearchCV
best_params = grid.best_params_
print("Hyperparameter terbaik:", best_params)
```

```python
# Lakukan evaluasi dengan beberapa metric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# Prediksi label dari data uji menggunakan model yang telah dituning
y_pred = grid.predict(X_test)

# Menampilkan laporan klasifikasi
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("")

# Menampilkan Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("")

# Menampilkan akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**PENJELASAN**

- Logistic Regression digunakan ketika kita ingin melakukan klasifikasi, seperti memprediksi apakah suatu observasi masuk ke dlm salah satu dari dua kelas (binary classification).

- Disini kita melakukan parameter tuning dg  GridSearchCV utk mencari kombinasi terbaik dari hyperparameter 'C' dan 'tol' yg meminimalkan error pada model Logistic Regression.

- Hasil akurasi model sekitar 0.8124, yang berarti model ini memiliki tingkat akurasi yang cukup baik dalam memprediksi label kelas.

- Model ini memiliki akurasi yg cukup baik dlm memprediksi kelas 0, tetapi memiliki masalah dlm memprediksi kelas 1, yang ditunjukkan oleh recall yang rendah dan F1-score yang rendah untuk kelas 1.

## Model 2 : Random Forest Classifier
Jelaskan secara Singkat Model ke-2 yang digunakan!

```python
from sklearn.ensemble import RandomForestClassifier

#START CODE

# Inisialisasi model Random Forest
model2 = RandomForestClassifier()

# Definisikan kumpulan hyperparameter yang akan diuji untuk model Random Forest
params = {
    'n_estimators': [50, 100, 200],  # Jumlah pohon dalam ensemble
    'max_depth': [5, 10, 20, 30],  # Kedalaman maksimum setiap pohon
    'min_samples_split': [2, 5, 10],  # Jumlah sampel minimum untuk membagi simpul internal
    'min_samples_leaf': [1, 2, 4],  # Jumlah sampel minimum di setiap daun pohon
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model2, # Model yang akan diuji
             param_grid= params, # Kumpulan hyperparameter yang akan diuji
             scoring = 'accuracy', # Metrik evaluasi yang digunakan
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 10-fold cross validation (artinya kita melakukan iterasi model sebanyak 10 kali)
            )

# Melatih model Random Forest dengan kombinasi hyperparameter terbaik
grid.fit(X_train, y_train)

# Menampilkan hyperparameter terbaik yang ditemukan oleh GridSearchCV untuk model Random Forest
best_params = grid.best_params_
print("Hyperparameter terbaik untuk Random Forest:", best_params)
```

```python
# Lakukan evaluasi dengan beberapa metric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# Prediksi label dari data uji menggunakan model yang telah dituning
y_pred = grid.predict(X_test)

# Menampilkan laporan klasifikasi
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("")

# Menampilkan Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("")

# Menampilkan akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**PENJELASAN**

- Model Random Forest yaitu Model ensemble yang menggabungkan beberapa pohon keputusan (decision trees) utk meningkatkan kinerja prediksi.

- Hyperparameter yg dgunakan = jmlh pohon (n_estimators), kedalaman maks setiap pohon (max_depth), jmlh sampel min utk membagi simpul internal (min_samples_split), dan jmlh sampel minimum di setiap daun pohon (min_samples_leaf).

- Parameter terbaik yang ditentukan memiliki akurasi yang lebih baik yaitu  0.8624 daripada model sebelumnya (Logistic Regression).

- Namun, model ini masih memiliki recall yang rendah untuk kelas 1 (terpenuhi), yang berarti masih perlu perbaikan dalam mengidentifikasi kasus positif kelas 1.


## Model 3 : Support Vector Machine (SVM)
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

#START CODE
# Inisialisasi model Support Vector Machine
model3 = SVC()

# Definisikan kumpulan hyperparameter yang akan diuji
params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Parameter kebalikan regulasi
    'kernel': ['rbf', 'poly', 'sigmoid'],  # Jenis kernel
    'gamma': [0.001, 0.01, 0.1, 1, 'scale', 'auto'],  # Parameter kernel
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model3, # Model yang akan diuji
             param_grid= params, # Kumpulan hyperparameter yang akan diuji
             scoring = 'accuracy', # Metrik evaluasi yang digunakan
             n_jobs = 10, # core cpu yang digunakan
             cv = 3 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )
# Melatih model dengan kombinasi hyperparameter terbaik
grid.fit(X_train, y_train)

# Menampilkan hyperparameter terbaik yang ditemukan oleh GridSearchCV
best_params = grid.best_params_
print("Hyperparameter terbaik:", best_params)
```

```python
# Lakukan evaluasi dengan beberapa metric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# Prediksi label dari data uji menggunakan model yang telah dituning
y_pred = grid.predict(X_test)

# Menampilkan laporan klasifikasi
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("")

# Menampilkan Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("")

# Menampilkan akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**PENJELASAN**

-  SVM digunakan utk mencari hyperplane terbaik yg memisahkan dua kelas dalam ruang fitur.

- Hyperparameter yg dgunakan =
<br> **C :** Nilai yang lebih tinggi mengarah pada margin yang lebih ketat.
<br> **kernel :** beberapa jenis kernel yg diuji -> 'rbf' (Radial Basis Function), 'poly' (Polynomial), dan 'sigmoid'.
<br> **gamma :** Parameter kernel yang mengatur seberapa jauh pengaruh dari satu titik data terhadap yang lain dalam fungsi kernel.

- Model SVM memiliki akurasi yg baik (0.8648) dlm memprediksi kelas mayoritas (kelas 0), tetapi memiliki performa yang lebih rendah dalam memprediksi kelas minoritas (kelas 1), seperti yang tercermin dari nilai recall dan F1-score yang lebih rendah untuk kelas 1.
